```python
path1 = "data_propercess/data/"
```

```python
#train 파일 목록가져오기
train_data = os.listdir(path1+"train")
```

```python
#train src,trg 분리
train_src_list =[]
train_trg_list =[]

for idx,texts in enumerate(train_data):
    if texts.split(".")[1] == "json":
        with open(path1+"train/"+texts, "r", encoding="utf-8") as file:
            train_json_data = json.load(file)
        text = train_json_data["utterance"]
        src_texts = [src['standard_form'] for src in text]
        for src in src_texts:
            train_src_list.append(src)
        trg_texts = [trg['dialect_form'] for trg in text]
        for trg in trg_texts:
            train_trg_list.append(trg)   
```

```python
print(len(train_src_list))
print(len(train_trg_list))
```

```python
# DataFrame_make

data_dict = {
    'src':[],
    'trg':[],}

for src,trg in zip(train_src_list,train_trg_list):
    hangul = re.compile('[^ㄱ-ㅣ가-힣\?! ]+') 
    src = hangul.sub("",src)
    trg = hangul.sub("",trg)
    data_dict['src'].append(src)
    data_dict['trg'].append(trg)

train_df = pd.DataFrame(data_dict)
train_df.head()

train_df.to_csv(path1+"train_df.csv", index=False)
```
